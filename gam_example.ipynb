# Simulate data with non-linear relationship

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def knn(x0, x, y, K=3):
    d = np.abs(x - x0)        
    idx = np.argsort(d)[:K]
    return np.mean(y[idx])

# Simulate data
n = 100
np.random.seed(123)
x1 = np.random.normal(3,2,size=n)
x2 = np.random.normal(3,1,size=n)
e = np.random.normal(0,0.1,size=n)
y = 1 + x1 - 0.6*x1**2 + 2*x2 + 0.6*x2**2 + e

# Prediction for (x1=2.5, x2=4)
# GAM, A few iterations
yd = y - y.mean()
fig, ax = plt.subplots(1,2)
ax[0].scatter(x1,yd)
ax[1].scatter(x2,yd)
```
